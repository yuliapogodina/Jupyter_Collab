1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'./laptop_price.csv', encoding='Windows-1251', index_col='laptop_ID')
df.sample(5)
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda row: row.split()[0])
df.sample(3)
```

1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)

```python
df['Memory'].value_counts()
```

```python
def convert_gb(row):
    mem = row.split()
    memory_out = 0
    for m in mem:
        if 'GB' in m:
            memory_out += float(m[:-2])
        elif 'TB' in m:
            memory_out += float(m[:-2])*1024
    return memory_out

convert_gb('256GB SSD +  1.0TB Hybrid')
```

```python
# 256GB SSD +  1.0TB Hybrid 
df['Memory_Amount'] = df['Memory'].apply(convert_gb)
df.sample(5)
```

1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)

```python
def memory_type(row):
    mem = row.split()
    type_m = ''
    for m in mem:
        if m.isalpha():
            if type_m:
                type_m += ' '
            type_m += m
    return type_m

memory_type('256GB SSD +  1.0TB Hybrid')
```

```python
df['Memory_Type'] = df['Memory'].apply(memory_type)
df.sample(15)
```

1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```
