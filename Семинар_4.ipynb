3.1 Постройте график в абсолютных величинах


```python
df.sample(2)
```

```python
data = pd.crosstab(index=df['Company'], columns=df['Cpu_Company']).drop(columns=['Samsung']).reset_index()
data
```

```python
plt.figure(figsize=(18, 6))
n_ticks = np.arange(len(data['AMD']))
offset = 0.2
w = 0.4
plt.bar(n_ticks - offset, data['AMD'], width=w)
plt.bar(n_ticks + offset, data['Intel'], width=w)

plt.xticks(n_ticks, data['Company'], rotation=30)
plt.title('Взаимосвязь CPu и компании производителя')
plt.xlabel('компания производитель ноутбуков')
plt.ylabel('Кол-во CPU')
plt.legend(['AMD', 'Intel']);
```

3.2 Постройте график в относительных величинах

```python
data_2 = pd.crosstab(index=df['Company'], columns=df['Cpu_Company'], normalize='index').drop(columns=['Samsung']).reset_index()
data_2
```

```python
plt.figure(figsize=(18, 6))

plt.bar(data_2['Company'], data_2['AMD'])
plt.bar(data_2['Company'], data_2['Intel'], bottom=data_2['AMD'])

plt.xticks(rotation=30)
plt.title('Взаимосвязь CPu и компании производителя (относительно)')
plt.xlabel('Компания производитель ноутбуков')
plt.ylabel('Кол-во CPU')
plt.legend(['AMD', 'Intel']);
```
